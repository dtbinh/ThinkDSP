1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a __different range of frequencies?__, and see what difference it makes.

The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
PI2 = 2 * numpy.pi
i = complex(0, 1)

def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = [0, 1, 2, 3] # frequencies 0 through N-1 
framerate = 11025
ts = numpy.linspace(0, 1, framerate)
```

```python
ys = synthesize2(amps, freqs, ts)
ys
```

```python
freqs2 = [4, 5, 6, 7] # frequencies N to 2N - 1
ys2 = synthesize2(amps, freqs2, ts)
ys2
```

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res
```

#### [1, -1] first-order difference approximation of first-derivative

```python
window = numpy.array([1, -1])
padded = zero_pad(window, 440)
fft_window = numpy.fft.rfft(padded)
thinkplot.plot(abs(fft_window))
```

#### [-1/2, 2, -3/2] second order difference approximation of first-derivative

```python
window = numpy.array([-1/2, 2, -3/2])
padded = zero_pad(window, 440)
fft_window = numpy.fft.rfft(padded)
thinkplot.plot(abs(fft_window))
```

#### [1, -2, 1] second-derivative approximation

```python
window = numpy.array([1, -2, 1])
# window = numpy.array([2, -5, 4,-1])
padded = zero_pad(window, 440)
fft_window = numpy.fft.rfft(padded)
thinkplot.plot(abs(fft_window))
```

The second derivative still looks high-pass in nature.  While not obvious, this is what we expect.  For the first-pass windows, as we convolve it over a signal, the areas of high inflection (large slope or change) will "activate" or "fire" values that are fire.  When I say this, convolving the window is simply doing the dot-product multiply of the window with the current elements.  

With the second-derivative approximation window [1, -2, 1], we can expect the critical points with sharp humps (points of deep convexity or concavity) to "activate" or "fire". This is because high values of the second-derivative indicate convexity or concavity.

Because points of deep convexity/concavity occurr as frequently as points of steep inflection, we can expect the first and second derivatives to act like high-pass filters in a similar way.

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
spectrum = thinkdsp.ParabolicSignal().make_wave().make_spectrum()
```

```python
spectrum.plot()
thinkplot.config(title='Parabolic Signal, Original')
```

```python
spectrum.differentiate()
```

```python
spectrum.plot()
thinkplot.config(title='Parabolic Signal with derivative filter applied')
```

As expected, the higher frequencies are amplified.  The differentiating filter is equivalent to a high-pass filter.
