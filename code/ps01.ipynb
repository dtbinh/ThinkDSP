###Question 6
A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats. See 

Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.  Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

```python
import scipy
from thinkdsp import PI2, unbias

class SawtoothSignal(thinkdsp.Signal):
    
    def __init__(self, freq=440, amp=1.0, offset=0):
        self.freq = freq
        self.amp = amp
        self.offset = offset
        self.func = scipy.signal.sawtooth
        
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        phases = PI2 * self.freq * ts + self.offset
        print 'phases: \n', phases
        ys = self.amp * self.func(phases)
        return ys
```

With the new ``SawtoothSignal`` class definition, I can make a sawtooth with a frequency of 3

```python
sawsig = SawtoothSignal(freq=3)
```

I can test the ``evaluate`` by using ``make_wave`` which calls ``evaluate``

```python
sawwave = sawsig.make_wave()
```

```python
sawwave.plot()
```

``SawtoothSignal`` success!  Within 1 second, 3 periods of the sawtooth is plotted as expected.

The harmonic frequencies of a sawtooth wave is unique in a couple ways:
* The frequency peaks are triangles, instead of tall skyscrapers.  What this means is that for a particular harmonic, there are frequencies around that harmonic that are still expressed.
