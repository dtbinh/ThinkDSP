```python
# copy the initial wave so we can apply different windowing 
wave_test = wave.copy()
wave_test.window(numpy.hamming(window_size))
wave_test.make_spectrum().plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
# copy the initial wave so we can apply different windowing 
wave_test = wave.copy()
wave_test.window(numpy.hanning(window_size))
wave_test.make_spectrum().plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
# copy the initial wave so we can apply different windowing 
wave_test = wave.copy()
wave_test.window(numpy.kaiser(window_size, beta=3))
wave_test.make_spectrum().plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

The windowing methods attempted all reduce the amplitude of the signal.  Each windowing method produces a different ouput shape of the spectrum.  The differences in shape are small, marginal, but still noticable.

2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

Hint: combine the evaluate functions from Chirp and SawtoothSignal.

Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
class SawtoothChirp(thinkdsp.Chirp):
    
    def evaluate(self, ts):
#         print('ts', ts)
        freqs = numpy.linspace(self.start, self.end, len(ts)-1)
#         print('freqs', freqs)
        dts = numpy.diff(ts)
#         print('dts', dts)
#         print('hi',thinkdsp.PI2 * freqs * dts)
        dphis = numpy.insert(thinkdsp.PI2 * freqs * dts, 0, 0)
#         print('dphis', dphis)
        phases = numpy.cumsum(dphis)
#         print('phases', phases)
        cycles = phases/thinkdsp.PI2
#         print('cycles', cycles)
        frac, _ = numpy.modf(cycles)
#         print('frac',frac)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
#         print('ys',ys)
        return ys

chirp = SawtoothChirp(start=440, end=440*6)
wave = chirp.make_wave(duration=5, framerate=11025)
spectogram = wave.make_spectrogram(seg_length=512)
spectogram.plot(high=2**8)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
wave.make_audio()
```

Interesting features of graph to take note
* at 4 seconds, there is a vertical white line.  These vertical white lines become less periodic as time goes on (from 0 - 5 seconds)
* There are many sloping lines because of the many harmonics used to express a sawtooth
* The _negative_ sloping lines are a product of aliasing.  These sloping lines are the same _positive_ sloping harmonics "folding" over at the cutoff frequency.

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
length = numpy.linspace(10, 80, 25)
pitch = 1. / length
thinkplot.Scatter(length, pitch)
thinkplot.Config(xlabel='time, which is proportional to trombone length ', ylabel='pitch')
```

The $1/f$ is more like a negative exponential chirp, as the pitch changes for a particular $\Delta t$ depends on the current $t$.

```python
class GlissandoChirp(thinkdsp.Chirp):
    """Represents a signal with varying frequency."""

    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        # express frequency limits in the time domain
        start, end = (1.0 / self.start, 1.0 / self.end)
        
        # express constant (linear) slide speed 
        slide = numpy.linspace(start, end, len(ts) - 1) 
        
        # invert to return to the frequency domain
        freqs = numpy.array(1.0 / slide, dtype=int)
        
        return self._evaluate(ts, freqs)
```

```python
dur = 3

# start C3 = 262
start = 262

# end F3 = 349
end = 349

trombone_up = GlissandoChirp(start=start, end=end).make_wave(duration=dur)
trombone_down = GlissandoChirp(start=end, end=start).make_wave(duration=dur)
trombone_up.apodize()
trombone_down.apodize()
up_down = trombone_up | trombone_down
spectogram = up_down.make_spectrogram(seg_length=512)
spectogram.plot(high=64)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', title='C3 to F3 Glissando', legend=False)
up_down.make_audio()
```

If the frequency change is within the same octave, the glissando looks and sounds more like a linear chirp.  However, we can see that it is indeed more like an exponential when we make the difference in start and end frequencies larger.

```python
dur = 3

# start C3 = 262
start = 262

# end F4 = 349*2
end = 349*2

trombone_up = GlissandoChirp(start=start, end=end).make_wave(duration=dur)
trombone_down = GlissandoChirp(start=end, end=start).make_wave(duration=dur)
trombone_up.apodize()
trombone_down.apodize()
up_down = trombone_up | trombone_down
spectogram = up_down.make_spectrogram(seg_length=512)
spectogram.plot(high=64)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', title='C3 to F4 Glissando', legend=False)
up_down.make_audio()
```

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
def barlett_method(wave, seg_length=4):
    """
    Arguments
    ---------
    wave: a Wave object
    seg_length: an integer representing number of seconds for each segment to be computed and averaged.
    
    Returns
    -------
    avg_power: average power array
    fs: the corresponding frequencies
    """
    num_segs = numpy.floor_divide(wave.duration, seg_length)
    print('number of segments or bins', num_segs)
    
    # xrange expects an integer argument
    duration = int(num_segs*seg_length)
    print('total duration (sec)', duration)
    
    # get the power spectra for all the segments
    powers = []
    for start in xrange(0, duration, seg_length):
        seg = wave.segment(start=start, duration=seg_length)
        spectrum = seg.make_spectrum()
        powers.append(spectrum.power)
    
    # compute the average power spectra
    avg_power = numpy.average(powers, axis=0)
    
    return avg_power, spectrum.fs
```

####Crickets...

```python
# Sorry for the semi-big upload.  But Bartlett's Method makes sense to use on "long signals"
wave = thinkdsp.read_wave(filename='wavs/22604__martypinso__dmp010037-crickets-texas.wav')

# account for half speed playback
wave.framerate *= 2

wave.make_audio()
```

```python
avg_power, fs = barlett_method(wave, seg_length=2)
```

```python
thinkplot.Plot(fs, avg_power, label='bartlett')
thinkplot.config(xlabel='frequency (Hz)', ylabel='power', xscale='log', yscale='log', title="Bartlett's Method for Crickets")
```

As mentioned by Wikipedia, the raw periodogram is not a good spectral estimate because of "the fact that the variance at a given frequency does not decrease as the number of samples used in the computation increases." The resulting graph indicates that Bartlett's Method does effectively smooth the spectral plot.
