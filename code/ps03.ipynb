2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

````python
# Functions taken from chap05.ipynb from the ThinkDSP github repository
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

def plot_autocorr(wave, label):
    """ Plots a simple autocorrelation of a wave 
    
    wave: Wave
    label: string
    """
    lags, corrs = autocorr(wave)
    thinkplot.plot(lags, corrs, label=label)
````

````python
plot_autocorr(wave, 'BitCoin')
thinkplot.config(xlabel='lag (index)', title='autocorrelation')
````

At about 85, the correlation coefficient is high

````python
def plot_shifted(wave, shift=0.002, start=0.2):
    """ plots the segment, and a shifted version of the segment, and computes the correlation between them """
    thinkplot.preplot(2)
    segment1 = wave.segment(start=start, duration=0.01)
    segment1.plot(linewidth=2, alpha=0.8)

    segment2 = wave.segment(start=start-shift, duration=0.01)
    segment2.plot(linewidth=2, alpha=0.4)

    corr = segment1.corr(segment2)
    text = r'$\rho =$ %.2g' % corr
    thinkplot.text(0.0005, -0.8, text)
    thinkplot.config(xlabel='time (s)', ylim=[-1, 1])
````

I used the interactive slider widget to find the shift in time that yielded the highest autocorrelation.

The result is below

````python
plot_shifted(segment, shift=.00215)
````

The shift that gives a $\rho = 1$ is equal to .00215.  This is the period (in seconds).

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

You will find code in ps03starter.ipynb that will help with this.

````python
from dct import analyze1, analyze2
import scipy
import timeit
````

### test for analyze1

````python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = (2) ** numpy.arange(4, 14)
print('ns',ns)

results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o analyze1(ys, freqs, ts)
    results.append(result)

# get the best run for the function
bests = [result.best for result in results]

# Compute Estimated Slope.  For polynomial time, this slope indicates the n^slope number
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)

# plot on a log-log scale
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', title='$O(n^%d)$'%t[0])
````
