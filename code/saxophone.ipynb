```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%matplotlib inline
```

Let's investigate the harmonic structure of a saxophone.  Here's a recording from freesound.org (thank you, iluppai).

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
wave.normalize()
wave.make_audio()
```

I'll pull out a segment with a constant pitch.

```python
start = 2.0
duration = 1.0
segment = wave.segment(start, duration)
segment.apodize()
segment.make_audio()
```

Here's what the waveform looks like:

```python
wave.segment(start, duration=0.03).plot()
thinkplot.config(xlabel='time (s)', ylim=[-1.05, 1.05], legend=False)
```

And here's the spectrum up to 2000 Hz:

```python
spectrum = segment.make_spectrum()
high = 2000
spectrum.plot(high=high)
```

Here are some of the peak frequencies and their amplitudes:

```python
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 500:
        print(f, amp)
```

The fundamental frequency is at 208 Hz, which is A$^\flat$, but the dominant frequency is one of the harmonics, at 623 Hz, which is the E$^\flat$ an octave and a fifth higher.

Nevertheless the pitch we perceive is the A$^\flat$.  For comparison, here's a triangle signal with the same fundamental frequency.

```python
wave = thinkdsp.TriangleSignal(freq=207.65).make_wave(duration=1)    # A flat
wave.apodize()
wave.make_audio()
```

Let's see what happens when we start knocking out harmonics.  Here's the original one more time:

```python
segment.make_audio()
```

I'll start by clobbering the fundamental:

```python
import copy
s = copy.deepcopy(spectrum)
s.band_stop(150, 250, 0.01)
s.plot(high=high)
```

Surprisingly, the perceived pitch is the same.  We still hear A$^\flat$, even though there is no component at that frequency.

```python
s.make_wave().make_audio()
```

The effect persists even if we clobber the next harmonic:

```python
s.band_stop(300, 500, 0.01)
s.plot(high=high)
```

Now the lowest frequency component is at 623 Hz, but we still hear 208 Hz.

```python
s.make_wave().make_audio()
```

This phenomenon is call the missing fundamental; you can read about it here. http://en.wikipedia.org/wiki/Missing_fundamental

Part of the explanation is that the higher harmonics contain information about what's missing.  If we clobber the high harmonics, the effect starts to fade:

```python
s.low_pass(1500, 0.01)
s.plot(high=high)
```

And when we clobber all of the high harmonics, the effect disappears and we hear a sinusoid at 623 Hz.

```python
s.low_pass(750, 0.01)
s.plot(high=2000)
```

```python
s.make_wave().make_audio()
```
